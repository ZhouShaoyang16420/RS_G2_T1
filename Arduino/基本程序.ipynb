```cpp
/* This minimal example shows how to get single-shot range
measurements from the VL6180X.

The range readings are in units of mm. */

#include <Wire.h>
#include <VL6180X.h>


VL6180X sensor;

double kp 
double ki 
double kd 

unsigned long currentTime, previousTime;
double elaspedTime;
double error;
double lastError;
double input,output,setPoint;
double cumError, rateError;

void setup() 
{
  Serial.begin(9600);
  Wire.begin();
  
  sensor.init();
  sensor.configureDefault();
  sensor.setTimeout(500);

  setPoint = 100;
  cumError = 0;
}

void loop() 
{ int val = sensor.readRangeSingleMillimeters();
  val = map(val, 0, 255,0,200);
  input = val;
  if ((input<=)
  {
    p=;
    i=;
    d=;
      }
      else if ((input>=)&&(input<=))
      {
        p=;
        i=;
        d=;
        }
        else(input>=)
        {
          p=;
          i=;
          d=;
          } 
  output = computePID(input);
  
  if (sensor.timeoutOccurred()) { Serial.print(" TIMEOUT"); }
  
   
  Serial.print(output);
  Serial.print(',');
  Serial.print(val);
  Serial.println();
  
}

double computePID(double input)
{
  currentTime =millis();
  elaspedTime = double(currentTime -previousTime);

  error = setPoint -input;
  cumError += error*elaspedTime;
  rateError = (error-lastError)/elaspedTime;

  double output = kp*error + ki*cumError + kd*rateError;

  lastError = error;
  previousTime = currentTime;

  return output; 
  } 
  ```
